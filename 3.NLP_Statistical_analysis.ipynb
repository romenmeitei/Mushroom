```python
# ============================================================
# NLP_Mushroom_2026_STATISTICS
# Statistical analysis notebook for mushroom-poisoning NLP study
# Input: CSV files generated from Stage I and Stage II notebooks
# ============================================================

!pip -q install pandas numpy scipy statsmodels scikit-posthocs openpyxl matplotlib

import os
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
from scipy.stats import (
    linregress, spearmanr, kendalltau, chi2_contingency,
    kruskal, mannwhitneyu
)
from statsmodels.stats.multitest import multipletests
import statsmodels.api as sm
import statsmodels.formula.api as smf

os.makedirs("statistics_outputs", exist_ok=True)
os.makedirs("statistics_outputs/tables", exist_ok=True)
os.makedirs("statistics_outputs/figures", exist_ok=True)

# ============================================================
# 1. UPLOAD CSV FILES
# ============================================================

uploaded = files.upload()

# Required / preferred files:
# 03_full_corpus_with_final_themes.csv
# 04_included_corpus_final_themes_only.csv
# 05_final_theme_counts.csv
# 06_final_theme_temporal_trends_long.csv
# 07_final_theme_temporal_trends_wide.csv
# 09_final_theme_relative_contribution.csv
# 27_species_toxin_keyword_frequency.csv
# 29_country_publication_counts.csv
# 15_topic_validation_metrics.csv

def find_file(possible_names):
    for name in possible_names:
        if name in uploaded.keys():
            return name
    for name in possible_names:
        if os.path.exists(name):
            return name
    return None

full_corpus_file = find_file([
    "03_full_corpus_with_final_themes.csv",
    "17_final_dataset_with_topics.csv"
])

included_file = find_file([
    "04_included_corpus_final_themes_only.csv",
    "03_full_corpus_with_final_themes.csv"
])

theme_counts_file = find_file([
    "05_final_theme_counts.csv",
    "18_final_topic_summary.csv"
])

trend_long_file = find_file([
    "06_final_theme_temporal_trends_long.csv",
    "19_temporal_trends_long.csv"
])

trend_wide_file = find_file([
    "07_final_theme_temporal_trends_wide.csv",
    "20_temporal_trends_wide.csv"
])

relative_file = find_file([
    "09_final_theme_relative_contribution.csv",
    "23_relative_topic_contribution.csv"
])

species_file = find_file([
    "27_species_toxin_keyword_frequency.csv"
])

country_file = find_file([
    "29_country_publication_counts.csv"
])

validation_file = find_file([
    "15_topic_validation_metrics.csv"
])

print("Files detected:")
print("Full corpus:", full_corpus_file)
print("Included corpus:", included_file)
print("Theme counts:", theme_counts_file)
print("Trend long:", trend_long_file)
print("Trend wide:", trend_wide_file)
print("Relative contribution:", relative_file)
print("Species/toxin:", species_file)
print("Country counts:", country_file)
print("Validation:", validation_file)

full_df = pd.read_csv(full_corpus_file) if full_corpus_file else None
included_df = pd.read_csv(included_file) if included_file else None
theme_counts = pd.read_csv(theme_counts_file) if theme_counts_file else None
trend_long = pd.read_csv(trend_long_file) if trend_long_file else None
trend_wide = pd.read_csv(trend_wide_file) if trend_wide_file else None
relative_df = pd.read_csv(relative_file) if relative_file else None
species_df = pd.read_csv(species_file) if species_file else None
country_df = pd.read_csv(country_file) if country_file else None
validation_df = pd.read_csv(validation_file) if validation_file else None

# ============================================================
# 2. STANDARDIZE COLUMNS
# ============================================================

def clean_columns(df):
    if df is None:
        return None
    df.columns = df.columns.str.strip()
    return df

full_df = clean_columns(full_df)
included_df = clean_columns(included_df)
theme_counts = clean_columns(theme_counts)
trend_long = clean_columns(trend_long)
trend_wide = clean_columns(trend_wide)
relative_df = clean_columns(relative_df)
species_df = clean_columns(species_df)
country_df = clean_columns(country_df)
validation_df = clean_columns(validation_df)

if included_df is not None and "year" in included_df.columns:
    included_df["year"] = pd.to_numeric(included_df["year"], errors="coerce")
    included_df = included_df.dropna(subset=["year"])
    included_df["year"] = included_df["year"].astype(int)

if trend_wide is not None:
    if "year" in trend_wide.columns:
        trend_wide["year"] = pd.to_numeric(trend_wide["year"], errors="coerce")
        trend_wide = trend_wide.dropna(subset=["year"])
        trend_wide["year"] = trend_wide["year"].astype(int)
        trend_wide = trend_wide.set_index("year")

# ============================================================
# 3. DESCRIPTIVE STATISTICS
# ============================================================

desc_rows = []

if full_df is not None:
    desc_rows.append(["Total corpus records", len(full_df)])

if included_df is not None:
    desc_rows.append(["Included expert-curated records", len(included_df)])
    if "final_theme" in included_df.columns:
        desc_rows.append(["Number of final themes", included_df["final_theme"].nunique()])
    if "year" in included_df.columns:
        desc_rows.append(["First publication year", included_df["year"].min()])
        desc_rows.append(["Last publication year", included_df["year"].max()])

if theme_counts is not None:
    count_col = "count" if "count" in theme_counts.columns else theme_counts.columns[1]
    desc_rows.append(["Largest theme count", theme_counts[count_col].max()])
    desc_rows.append(["Smallest theme count", theme_counts[count_col].min()])

descriptive_stats = pd.DataFrame(desc_rows, columns=["metric", "value"])
descriptive_stats.to_csv("statistics_outputs/tables/01_descriptive_statistics.csv", index=False)
display(descriptive_stats)

# ============================================================
# 4. THEME DISTRIBUTION STATISTICS
# Chi-square goodness-of-fit against equal distribution
# ============================================================

if theme_counts is not None:
    if "final_theme" not in theme_counts.columns:
        theme_counts.columns = ["final_theme", "count"] + list(theme_counts.columns[2:])

    count_col = "count"
    observed = theme_counts[count_col].values
    expected = np.repeat(observed.sum() / len(observed), len(observed))

    chi2_stat = np.sum((observed - expected) ** 2 / expected)
    df_chi = len(observed) - 1

    from scipy.stats import chi2
    chi2_p = 1 - chi2.cdf(chi2_stat, df_chi)

    chi_df = pd.DataFrame({
        "test": ["Chi-square goodness-of-fit"],
        "null_hypothesis": ["Final themes are equally represented"],
        "chi_square": [chi2_stat],
        "df": [df_chi],
        "p_value": [chi2_p]
    })

    chi_df.to_csv("statistics_outputs/tables/02_theme_distribution_chisquare.csv", index=False)
    display(chi_df)

# ============================================================
# 5. ANNUAL PUBLICATION TREND: TOTAL CORPUS
# Linear regression + Poisson regression + Spearman
# ============================================================

if included_df is not None and "year" in included_df.columns:
    annual_counts = (
        included_df.groupby("year")
        .size()
        .reset_index(name="publications")
        .sort_values("year")
    )

    x = annual_counts["year"].values
    y = annual_counts["publications"].values

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    rho, rho_p = spearmanr(x, y)
    tau, tau_p = kendalltau(x, y)

    # Poisson regression
    poisson_model = smf.glm(
        formula="publications ~ year",
        data=annual_counts,
        family=sm.families.Poisson()
    ).fit()

    annual_stats = pd.DataFrame({
        "analysis": [
            "Linear regression",
            "Spearman correlation",
            "Kendall tau",
            "Poisson regression"
        ],
        "statistic": [
            slope,
            rho,
            tau,
            poisson_model.params["year"]
        ],
        "r_squared": [
            r_value**2,
            np.nan,
            np.nan,
            np.nan
        ],
        "p_value": [
            p_value,
            rho_p,
            tau_p,
            poisson_model.pvalues["year"]
        ],
        "interpretation": [
            "Annual change in number of publications",
            "Monotonic association between year and publication count",
            "Rank-based temporal association",
            "Log-rate change in publications per year"
        ]
    })

    annual_counts.to_csv("statistics_outputs/tables/03_annual_publication_counts.csv", index=False)
    annual_stats.to_csv("statistics_outputs/tables/04_annual_publication_trend_statistics.csv", index=False)

    display(annual_counts.head())
    display(annual_stats)

    plt.figure(figsize=(8, 5))
    plt.scatter(x, y)
    plt.plot(x, intercept + slope * x)
    plt.xlabel("Year")
    plt.ylabel("Number of publications")
    plt.title("Annual publication trend in mushroom-poisoning literature")
    plt.tight_layout()
    plt.savefig("statistics_outputs/figures/01_annual_publication_trend.png", dpi=300)
    plt.show()

# ============================================================
# 6. THEME-SPECIFIC TREND STATISTICS
# Linear, Spearman, Kendall, Poisson regression
# Multiple-testing correction
# ============================================================

if trend_wide is not None:
    trend_results = []

    for theme in trend_wide.columns:
        y = trend_wide[theme].values.astype(float)
        x = trend_wide.index.values.astype(float)

        if np.sum(y) == 0 or len(np.unique(y)) <= 1:
            continue

        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        rho, rho_p = spearmanr(x, y)
        tau, tau_p = kendalltau(x, y)

        tmp = pd.DataFrame({"year": x, "count": y})

        try:
            poisson = smf.glm(
                formula="count ~ year",
                data=tmp,
                family=sm.families.Poisson()
            ).fit()
            poisson_beta = poisson.params["year"]
            poisson_p = poisson.pvalues["year"]
        except Exception:
            poisson_beta = np.nan
            poisson_p = np.nan

        trend_results.append({
            "final_theme": theme,
            "total_publications": y.sum(),
            "linear_slope_publications_per_year": slope,
            "linear_r_squared": r_value**2,
            "linear_p_value": p_value,
            "spearman_rho": rho,
            "spearman_p_value": rho_p,
            "kendall_tau": tau,
            "kendall_p_value": tau_p,
            "poisson_beta_log_rate_per_year": poisson_beta,
            "poisson_p_value": poisson_p
        })

    theme_trend_stats = pd.DataFrame(trend_results)

    for pcol in ["linear_p_value", "spearman_p_value", "kendall_p_value", "poisson_p_value"]:
        mask = theme_trend_stats[pcol].notna()
        corrected = np.full(len(theme_trend_stats), np.nan)
        if mask.sum() > 0:
            corrected[mask] = multipletests(theme_trend_stats.loc[mask, pcol], method="fdr_bh")[1]
        theme_trend_stats[pcol.replace("p_value", "q_value_fdr")] = corrected

    theme_trend_stats = theme_trend_stats.sort_values("linear_p_value")
    theme_trend_stats.to_csv("statistics_outputs/tables/05_theme_specific_trend_statistics_fdr.csv", index=False)
    display(theme_trend_stats)

# ============================================================
# 7. PERIOD COMPARISON
# Compare pre-2010, 2010–2017, 2018–2025
# ============================================================

if included_df is not None and "year" in included_df.columns and "final_theme" in included_df.columns:

    def period_label(year):
        if year <= 2009:
            return "2000–2009"
        elif year <= 2017:
            return "2010–2017"
        else:
            return "2018–2025"

    included_df["period"] = included_df["year"].apply(period_label)

    period_theme = (
        included_df.groupby(["period", "final_theme"])
        .size()
        .reset_index(name="count")
    )

    period_wide = (
        period_theme.pivot(index="period", columns="final_theme", values="count")
        .fillna(0)
    )

    period_theme.to_csv("statistics_outputs/tables/06_period_theme_counts_long.csv", index=False)
    period_wide.reset_index().to_csv("statistics_outputs/tables/07_period_theme_counts_wide.csv", index=False)

    contingency = period_wide.values
    chi2_stat, chi2_p, dof, expected = chi2_contingency(contingency)

    period_chi = pd.DataFrame({
        "test": ["Chi-square test of independence"],
        "comparison": ["Publication period × final theme"],
        "chi_square": [chi2_stat],
        "df": [dof],
        "p_value": [chi2_p]
    })

    period_chi.to_csv("statistics_outputs/tables/08_period_theme_chisquare.csv", index=False)

    expected_df = pd.DataFrame(expected, index=period_wide.index, columns=period_wide.columns)
    expected_df.to_csv("statistics_outputs/tables/09_period_theme_expected_counts.csv")

    display(period_wide)
    display(period_chi)

# ============================================================
# 8. THEME ENRICHMENT IN RECENT PERIOD
# Odds ratio for 2018–2025 vs 2000–2017
# ============================================================

if included_df is not None and "year" in included_df.columns and "final_theme" in included_df.columns:

    included_df["recent_period"] = np.where(included_df["year"] >= 2018, "2018–2025", "2000–2017")

    enrichment_rows = []
    themes = sorted(included_df["final_theme"].dropna().unique())

    for theme in themes:
        a = len(included_df[(included_df["final_theme"] == theme) & (included_df["recent_period"] == "2018–2025")])
        b = len(included_df[(included_df["final_theme"] != theme) & (included_df["recent_period"] == "2018–2025")])
        c = len(included_df[(included_df["final_theme"] == theme) & (included_df["recent_period"] == "2000–2017")])
        d = len(included_df[(included_df["final_theme"] != theme) & (included_df["recent_period"] == "2000–2017")])

        # Haldane-Anscombe correction
        ah, bh, ch, dh = a + 0.5, b + 0.5, c + 0.5, d + 0.5
        odds_ratio = (ah * dh) / (bh * ch)
        se_log_or = np.sqrt(1/ah + 1/bh + 1/ch + 1/dh)
        ci_low = np.exp(np.log(odds_ratio) - 1.96 * se_log_or)
        ci_high = np.exp(np.log(odds_ratio) + 1.96 * se_log_or)

        chi2_stat, p_val, _, _ = chi2_contingency([[a, b], [c, d]])

        enrichment_rows.append({
            "final_theme": theme,
            "recent_count_2018_2025": a,
            "earlier_count_2000_2017": c,
            "odds_ratio_recent_vs_earlier": odds_ratio,
            "ci95_low": ci_low,
            "ci95_high": ci_high,
            "p_value": p_val
        })

    enrichment_df = pd.DataFrame(enrichment_rows)
    enrichment_df["q_value_fdr"] = multipletests(enrichment_df["p_value"], method="fdr_bh")[1]
    enrichment_df = enrichment_df.sort_values("odds_ratio_recent_vs_earlier", ascending=False)

    enrichment_df.to_csv("statistics_outputs/tables/10_recent_period_theme_enrichment.csv", index=False)
    display(enrichment_df)

# ============================================================
# 9. DIVERSITY ANALYSIS OVER TIME
# Shannon diversity annually and by period
# ============================================================

def shannon_from_counts(counts):
    counts = np.array(counts, dtype=float)
    counts = counts[counts > 0]
    if counts.sum() == 0:
        return np.nan
    p = counts / counts.sum()
    return -np.sum(p * np.log(p))

def simpson_from_counts(counts):
    counts = np.array(counts, dtype=float)
    counts = counts[counts > 0]
    if counts.sum() == 0:
        return np.nan
    p = counts / counts.sum()
    return 1 - np.sum(p ** 2)

if trend_wide is not None:
    diversity_rows = []

    for year, row in trend_wide.iterrows():
        counts = row.values
        diversity_rows.append({
            "year": year,
            "total_publications": np.sum(counts),
            "shannon_H": shannon_from_counts(counts),
            "simpson_1_minus_D": simpson_from_counts(counts),
            "theme_richness": np.sum(np.array(counts) > 0)
        })

    diversity_year = pd.DataFrame(diversity_rows)

    x = diversity_year["year"].values
    y = diversity_year["shannon_H"].values
    mask = ~np.isnan(y)

    if mask.sum() > 3:
        slope, intercept, r_value, p_value, std_err = linregress(x[mask], y[mask])
        rho, rho_p = spearmanr(x[mask], y[mask])
    else:
        slope = intercept = r_value = p_value = std_err = rho = rho_p = np.nan

    diversity_trend = pd.DataFrame({
        "metric": ["Annual Shannon diversity"],
        "linear_slope_per_year": [slope],
        "linear_r_squared": [r_value**2 if not np.isnan(r_value) else np.nan],
        "linear_p_value": [p_value],
        "spearman_rho": [rho],
        "spearman_p_value": [rho_p]
    })

    diversity_year.to_csv("statistics_outputs/tables/11_annual_thematic_diversity.csv", index=False)
    diversity_trend.to_csv("statistics_outputs/tables/12_annual_diversity_trend_statistics.csv", index=False)

    display(diversity_year.head())
    display(diversity_trend)

    plt.figure(figsize=(8, 5))
    plt.plot(diversity_year["year"], diversity_year["shannon_H"], marker="o")
    plt.xlabel("Year")
    plt.ylabel("Shannon diversity index")
    plt.title("Annual thematic diversity of mushroom-poisoning research")
    plt.tight_layout()
    plt.savefig("statistics_outputs/figures/02_annual_shannon_diversity.png", dpi=300)
    plt.show()

# ============================================================
# 10. PRE-POST COMPARISON FOR AI THEME
# Checks whether AI-assisted identification increased after 2018
# ============================================================

if included_df is not None and "final_theme" in included_df.columns and "year" in included_df.columns:
    ai_mask = included_df["final_theme"].astype(str).str.contains("AI-assisted|computational", case=False, regex=True)
    included_df["ai_theme"] = np.where(ai_mask, "AI/computational", "Other themes")
    included_df["post_2018"] = np.where(included_df["year"] >= 2018, "2018–2025", "2000–2017")

    ai_table = pd.crosstab(included_df["post_2018"], included_df["ai_theme"])
    chi2_stat, chi2_p, dof, expected = chi2_contingency(ai_table)

    ai_stats = pd.DataFrame({
        "test": ["Chi-square test of independence"],
        "comparison": ["AI/computational theme enrichment after 2018"],
        "chi_square": [chi2_stat],
        "df": [dof],
        "p_value": [chi2_p]
    })

    ai_table.to_csv("statistics_outputs/tables/13_ai_theme_pre_post_counts.csv")
    ai_stats.to_csv("statistics_outputs/tables/14_ai_theme_pre_post_chisquare.csv", index=False)

    display(ai_table)
    display(ai_stats)

# ============================================================
# 11. SPECIES / TOXIN FREQUENCY STATISTICS
# ============================================================

if species_df is not None:
    if "term" not in species_df.columns:
        species_df.columns = ["term", "count"] + list(species_df.columns[2:])

    species_df["rank"] = species_df["count"].rank(ascending=False, method="dense").astype(int)

    total_mentions = species_df["count"].sum()
    species_df["share_of_species_toxin_mentions_percent"] = (species_df["count"] / total_mentions * 100).round(2)

    observed = species_df["count"].values
    expected = np.repeat(observed.sum() / len(observed), len(observed))
    chi2_stat = np.sum((observed - expected) ** 2 / expected)
    df_chi = len(observed) - 1
    from scipy.stats import chi2
    chi2_p = 1 - chi2.cdf(chi2_stat, df_chi)

    species_chi = pd.DataFrame({
        "test": ["Chi-square goodness-of-fit"],
        "null_hypothesis": ["Species/toxin terms are equally frequent"],
        "chi_square": [chi2_stat],
        "df": [df_chi],
        "p_value": [chi2_p]
    })

    species_df.to_csv("statistics_outputs/tables/15_species_toxin_ranked_frequency.csv", index=False)
    species_chi.to_csv("statistics_outputs/tables/16_species_toxin_frequency_chisquare.csv", index=False)

    display(species_df)
    display(species_chi)

# ============================================================
# 12. COUNTRY CONTRIBUTION STATISTICS
# ============================================================

if country_df is not None:
    if "country" not in country_df.columns:
        country_df.columns = ["country", "count"] + list(country_df.columns[2:])

    country_df["rank"] = country_df["count"].rank(ascending=False, method="dense").astype(int)
    country_df["percent"] = (country_df["count"] / country_df["count"].sum() * 100).round(2)

    top10 = country_df.sort_values("count", ascending=False).head(10)
    top10_share = top10["count"].sum() / country_df["count"].sum() * 100

    country_summary = pd.DataFrame({
        "metric": ["Number of countries", "Top 10 countries share (%)", "Largest country contribution"],
        "value": [country_df["country"].nunique(), top10_share, country_df["count"].max()]
    })

    country_df.to_csv("statistics_outputs/tables/17_country_ranked_publication_counts.csv", index=False)
    country_summary.to_csv("statistics_outputs/tables/18_country_contribution_summary.csv", index=False)

    display(country_df.head(20))
    display(country_summary)

# ============================================================
# 13. TOPIC VALIDATION METRICS SUMMARY
# ============================================================

if validation_df is not None:
    validation_df.to_csv("statistics_outputs/tables/19_topic_validation_metrics_summary.csv", index=False)
    display(validation_df)

# ============================================================
# 14. MANUSCRIPT-READY STATISTICAL SUMMARY TEXT
# ============================================================

lines = []

if 'annual_stats' in globals():
    lin = annual_stats[annual_stats["analysis"] == "Linear regression"].iloc[0]
    lines.append(
        f"Overall annual publication output increased over time "
        f"(linear slope = {lin['statistic']:.3f} publications/year, "
        f"R² = {lin['r_squared']:.3f}, p = {lin['p_value']:.3g})."
    )

if 'theme_trend_stats' in globals():
    sig = theme_trend_stats[
        theme_trend_stats["linear_q_value_fdr"] < 0.05
    ].copy()
    lines.append(
        f"Theme-specific trend analysis identified {len(sig)} final themes with "
        f"statistically significant temporal increases after FDR correction."
    )

if 'period_chi' in globals():
    row = period_chi.iloc[0]
    lines.append(
        f"The distribution of final themes differed significantly across publication periods "
        f"(χ² = {row['chi_square']:.2f}, df = {int(row['df'])}, p = {row['p_value']:.3g})."
    )

if 'diversity_trend' in globals():
    row = diversity_trend.iloc[0]
    lines.append(
        f"Annual thematic diversity showed a temporal trend "
        f"(Shannon slope = {row['linear_slope_per_year']:.4f}/year, "
        f"R² = {row['linear_r_squared']:.3f}, p = {row['linear_p_value']:.3g})."
    )

if 'ai_stats' in globals():
    row = ai_stats.iloc[0]
    lines.append(
        f"AI/computational identification themes showed period-associated enrichment "
        f"(χ² = {row['chi_square']:.2f}, df = {int(row['df'])}, p = {row['p_value']:.3g})."
    )

stat_text = "\n".join(lines)

with open("statistics_outputs/tables/20_manuscript_ready_statistics_text.txt", "w") as f:
    f.write(stat_text)

print(stat_text)

# ============================================================
# 15. ZIP OUTPUTS
# ============================================================

zip_path = "mushroom_poisoning_statistics_outputs.zip"

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    for root, dirs, file_list in os.walk("statistics_outputs"):
        for file in file_list:
            full_path = os.path.join(root, file)
            z.write(full_path)

print("Created:", zip_path)
files.download(zip_path)
```
