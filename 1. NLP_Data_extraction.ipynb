```python
# ============================================================
# PUBLICATION-READY NLP / BIBLIOMETRIC PIPELINE
# TOPIC: Mushroom poisoning / mycetism / amatoxin literature
# Sources: PubMed + Scopus
# Validation subset: Web of Science Starter API
# Outputs: cleaned corpus, PRISMA counts, BERTopic topics,
# coherence metrics, trend statistics, manuscript-ready figures
# ============================================================

# ============================================================
# CELL 1. INSTALL PACKAGES
# ============================================================

!pip -q install pandas numpy requests tqdm biopython rapidfuzz openpyxl \
    bertopic umap-learn hdbscan sentence-transformers gensim pycountry \
    matplotlib scipy scikit-learn plotly kaleido

# ============================================================
# CELL 2. IMPORTS AND GLOBAL SETTINGS
# ============================================================

import os
import re
import time
import json
import math
import zipfile
import warnings
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm.auto import tqdm
from Bio import Entrez
from rapidfuzz import fuzz
from scipy.stats import linregress
from google.colab import userdata, files

from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN

from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_colwidth", 300)

os.makedirs("outputs", exist_ok=True)
os.makedirs("outputs/figures", exist_ok=True)
os.makedirs("outputs/tables", exist_ok=True)
os.makedirs("outputs/raw", exist_ok=True)

# ============================================================
# CELL 3. SECURE API CONFIGURATION
# ============================================================

# Add these in Colab:
# Left panel → 🔑 Secrets → Add:
# SCOPUS_API_KEY
# WOS_API_KEY
# ENTREZ_EMAIL

SCOPUS_API_KEY = userdata.get("SCOPUS_API_KEY")
WOS_API_KEY = userdata.get("WOS_API_KEY")
ENTREZ_EMAIL = userdata.get("ENTREZ_EMAIL")

ENTREZ_TOOL = "mushroom_poisoning_nlp_pipeline"

assert ENTREZ_EMAIL is not None, "Missing ENTREZ_EMAIL in Colab Secrets."
assert SCOPUS_API_KEY is not None, "Missing SCOPUS_API_KEY in Colab Secrets."

Entrez.email = ENTREZ_EMAIL
Entrez.tool = ENTREZ_TOOL

YEAR_START = 2000
YEAR_END = 2025

SCOPUS_QUERY = (
    'TITLE-ABS-KEY("mushroom poisoning" OR "mushroom intoxication" OR '
    '"poisonous mushroom" OR mycetism OR amatoxin OR amanitin OR '
    '"mushroom toxin" OR "Amanita poisoning") '
    f'AND PUBYEAR > {YEAR_START - 1} AND PUBYEAR < {YEAR_END + 1}'
)

PUBMED_QUERY = f"""
(
  "mushroom poisoning"[Title/Abstract]
  OR "mushroom intoxication"[Title/Abstract]
  OR "poisonous mushroom"[Title/Abstract]
  OR mycetism[Title/Abstract]
  OR amatoxin[Title/Abstract]
  OR amanitin[Title/Abstract]
  OR "mushroom toxin"[Title/Abstract]
  OR "Amanita poisoning"[Title/Abstract]
)
AND ("{YEAR_START}/01/01"[Date - Publication] : "{YEAR_END}/12/31"[Date - Publication])
AND english[Language]
"""

WOS_QUERY = (
    'TS=("mushroom poisoning" OR "mushroom intoxication" OR '
    '"poisonous mushroom" OR mycetism OR amatoxin OR amanitin '
    'OR "mushroom toxin" OR "Amanita poisoning")'
)

SCOPUS_MAX_RECORDS = 5000
SCOPUS_BATCH_SIZE = 25
PUBMED_BATCH_SIZE = 200
WOS_MAX_RECORDS = 300
WOS_LIMIT_PER_PAGE = 50

REQUIRE_ABSTRACT_FOR_NLP = False
TITLE_SIM_THRESHOLD = 96

print("Configuration loaded safely.")

# ============================================================
# CELL 4. HELPER FUNCTIONS
# ============================================================

def clean_text(x):
    if x is None or pd.isna(x):
        return ""
    x = str(x)
    x = re.sub(r"<[^>]+>", " ", x)
    x = re.sub(r"\s+", " ", x)
    return x.strip()

def normalize_title(x):
    x = clean_text(x).lower()
    x = re.sub(r"[^a-z0-9\s]", " ", x)
    x = re.sub(r"\s+", " ", x).strip()
    return x

def normalize_doi(x):
    if x is None or pd.isna(x):
        return ""
    x = str(x).strip().lower()
    x = x.replace("https://doi.org/", "").replace("http://doi.org/", "")
    x = re.sub(r"\s+", "", x)
    return x

def extract_year(x):
    if x is None or pd.isna(x):
        return np.nan
    m = re.search(r"(19|20)\d{2}", str(x))
    return int(m.group()) if m else np.nan

def safe_join(values):
    values = [clean_text(v) for v in values if clean_text(v)]
    return "; ".join(sorted(set(values)))

def save_csv(df, name):
    path = f"outputs/tables/{name}"
    df.to_csv(path, index=False)
    print(f"Saved: {path} | shape={df.shape}")
    return path

# ============================================================
# CELL 5. PUBMED RETRIEVAL
# ============================================================

def fetch_pubmed_ids(query, max_records=100000):
    handle = Entrez.esearch(
        db="pubmed",
        term=query,
        retmax=max_records,
        sort="pub date"
    )
    result = Entrez.read(handle)
    handle.close()
    return result.get("IdList", [])

def parse_pubmed_article(article):
    medline = article.get("MedlineCitation", {})
    pmid = str(medline.get("PMID", ""))

    art = medline.get("Article", {})
    title = clean_text(art.get("ArticleTitle", ""))

    abstract_parts = []
    abst = art.get("Abstract", {})
    if "AbstractText" in abst:
        for part in abst["AbstractText"]:
            abstract_parts.append(clean_text(part))
    abstract = clean_text(" ".join(abstract_parts))

    journal = art.get("Journal", {})
    journal_title = clean_text(journal.get("Title", ""))
    pub_date = ""

    try:
        pub_date_dict = journal.get("JournalIssue", {}).get("PubDate", {})
        pub_date = " ".join([str(v) for v in pub_date_dict.values()])
    except Exception:
        pass

    year = extract_year(pub_date)

    doi = ""
    article_ids = article.get("PubmedData", {}).get("ArticleIdList", [])
    for aid in article_ids:
        if str(getattr(aid, "attributes", {}).get("IdType", "")).lower() == "doi":
            doi = normalize_doi(str(aid))

    authors = []
    affiliations = []

    for a in art.get("AuthorList", []):
        last = clean_text(a.get("LastName", ""))
        fore = clean_text(a.get("ForeName", ""))
        name = clean_text(f"{fore} {last}")
        if name:
            authors.append(name)

        for aff in a.get("AffiliationInfo", []):
            affiliations.append(clean_text(aff.get("Affiliation", "")))

    keywords = []
    for kwlist in medline.get("KeywordList", []):
        for kw in kwlist:
            keywords.append(clean_text(str(kw)))

    return {
        "source": "PubMed",
        "pmid": pmid,
        "doi": doi,
        "eid": "",
        "title": title,
        "abstract": abstract,
        "year": year,
        "journal": journal_title,
        "authors": safe_join(authors),
        "affiliations": safe_join(affiliations),
        "keywords": safe_join(keywords),
        "language": "English",
        "document_type": "Journal Article"
    }

def fetch_pubmed_records(query):
    ids = fetch_pubmed_ids(query)
    print(f"PubMed IDs retrieved: {len(ids)}")

    records = []
    for i in tqdm(range(0, len(ids), PUBMED_BATCH_SIZE)):
        batch = ids[i:i + PUBMED_BATCH_SIZE]
        handle = Entrez.efetch(
            db="pubmed",
            id=",".join(batch),
            retmode="xml"
        )
        data = Entrez.read(handle)
        handle.close()

        for article in data.get("PubmedArticle", []):
            records.append(parse_pubmed_article(article))

        time.sleep(0.34)

    df = pd.DataFrame(records)
    return df

pubmed_df = fetch_pubmed_records(PUBMED_QUERY)
save_csv(pubmed_df, "01_raw_pubmed.csv")
pubmed_df.head()

# ============================================================
# CELL 6. SCOPUS RETRIEVAL
# ============================================================

# ============================================================
# REPLACE CELL 6. SCOPUS RETRIEVAL WITH THIS
# ============================================================

def empty_literature_df(source_name="Scopus"):
    return pd.DataFrame(columns=[
        "source", "pmid", "doi", "eid", "title", "abstract", "year",
        "journal", "authors", "affiliations", "keywords",
        "language", "document_type"
    ])

def scopus_search(query):
    url = "https://api.elsevier.com/content/search/scopus"
    headers = {
        "X-ELS-APIKey": SCOPUS_API_KEY,
        "Accept": "application/json"
    }

    records = []

    # Use STANDARD view instead of COMPLETE.
    # COMPLETE often causes 401 authorization errors.
    for start in tqdm(range(0, SCOPUS_MAX_RECORDS, SCOPUS_BATCH_SIZE)):
        params = {
            "query": query,
            "start": start,
            "count": SCOPUS_BATCH_SIZE,
            "view": "STANDARD"
        }

        r = requests.get(url, headers=headers, params=params, timeout=30)

        if r.status_code != 200:
            print("Scopus error:", r.status_code, r.text[:500])
            print("Continuing with PubMed-only corpus.")
            return empty_literature_df("Scopus")

        data = r.json()
        entries = data.get("search-results", {}).get("entry", [])

        if not entries:
            break

        for e in entries:
            records.append({
                "source": "Scopus",
                "pmid": clean_text(e.get("pubmed-id", "")),
                "doi": normalize_doi(e.get("prism:doi", "")),
                "eid": clean_text(e.get("eid", "")),
                "title": clean_text(e.get("dc:title", "")),
                "abstract": clean_text(e.get("dc:description", "")),
                "year": extract_year(e.get("prism:coverDate", "")),
                "journal": clean_text(e.get("prism:publicationName", "")),
                "authors": clean_text(e.get("dc:creator", "")),
                "affiliations": "",
                "keywords": "",
                "language": "",
                "document_type": clean_text(e.get("subtypeDescription", ""))
            })

        total = int(data.get("search-results", {}).get("opensearch:totalResults", 0))

        if start + SCOPUS_BATCH_SIZE >= total:
            break

        time.sleep(0.25)

    if len(records) == 0:
        return empty_literature_df("Scopus")

    return pd.DataFrame(records)

scopus_df = scopus_search(SCOPUS_QUERY)
save_csv(scopus_df, "02_raw_scopus.csv")
scopus_df.head()
# ============================================================
# CELL 7. OPTIONAL: FETCH MISSING SCOPUS ABSTRACTS BY EID
# ============================================================

# ============================================================
# REPLACE CELL 7. OPTIONAL SCOPUS ABSTRACT FETCHING WITH THIS
# ============================================================

def fetch_scopus_abstract_by_eid(eid):
    if not eid:
        return ""

    url = f"https://api.elsevier.com/content/abstract/eid/{eid}"
    headers = {
        "X-ELS-APIKey": SCOPUS_API_KEY,
        "Accept": "application/json"
    }

    try:
        r = requests.get(url, headers=headers, timeout=30)

        if r.status_code != 200:
            return ""

        data = r.json()
        core = data.get("abstracts-retrieval-response", {}).get("coredata", {})
        return clean_text(core.get("dc:description", ""))

    except Exception:
        return ""

if len(scopus_df) == 0:
    print("Scopus returned no records. Skipping Scopus abstract retrieval.")
else:
    missing_abs = scopus_df["abstract"].fillna("").str.len() < 30
    print("Scopus records with missing/short abstracts:", missing_abs.sum())

    for idx in tqdm(scopus_df[missing_abs].index[:1000]):
        eid = scopus_df.loc[idx, "eid"]
        abs_text = fetch_scopus_abstract_by_eid(eid)

        if abs_text:
            scopus_df.loc[idx, "abstract"] = abs_text

        time.sleep(0.2)

save_csv(scopus_df, "03_raw_scopus_with_abstract_attempt.csv")

# ============================================================
# CELL 8. OPTIONAL WOS STARTER VALIDATION RETRIEVAL
# ============================================================

def fetch_wos_validation(query):
    if WOS_API_KEY is None:
        print("No WOS_API_KEY found. Skipping WoS validation retrieval.")
        return pd.DataFrame()

    headers = {
        "X-ApiKey": WOS_API_KEY,
        "Accept": "application/json"
    }

    records = []
    base = "https://api.clarivate.com/apis/wos-starter/v1/documents"

    for first in tqdm(range(1, WOS_MAX_RECORDS + 1, WOS_LIMIT_PER_PAGE)):
        params = {
            "q": query,
            "db": "WOS",
            "limit": WOS_LIMIT_PER_PAGE,
            "page": int((first - 1) / WOS_LIMIT_PER_PAGE) + 1
        }

        try:
            r = requests.get(base, headers=headers, params=params, timeout=30)
            if r.status_code != 200:
                print("WoS error:", r.status_code, r.text[:500])
                break

            data = r.json()
            hits = data.get("hits", [])

            if not hits:
                break

            for h in hits:
                names = h.get("names", {})
                authors = names.get("authors", [])
                author_names = []
                for a in authors:
                    if isinstance(a, dict):
                        author_names.append(clean_text(a.get("displayName", "")))

                source = h.get("source", {})
                identifiers = h.get("identifiers", {})

                records.append({
                    "source": "WoS_validation",
                    "pmid": "",
                    "doi": normalize_doi(identifiers.get("doi", "")),
                    "eid": "",
                    "title": clean_text(h.get("title", "")),
                    "abstract": clean_text(h.get("abstract", "")),
                    "year": extract_year(h.get("sourcePublishYear", "")),
                    "journal": clean_text(source.get("sourceTitle", "")),
                    "authors": safe_join(author_names),
                    "affiliations": "",
                    "keywords": safe_join(h.get("keywords", [])) if isinstance(h.get("keywords", []), list) else "",
                    "language": "",
                    "document_type": clean_text(h.get("documentType", ""))
                })

            time.sleep(0.4)

        except Exception as e:
            print("WoS exception:", e)
            break

    return pd.DataFrame(records)

wos_df = fetch_wos_validation(WOS_QUERY)
save_csv(wos_df, "04_raw_wos_validation.csv")
wos_df.head()

# ============================================================
# CELL 9. MERGE PUBMED + SCOPUS
# ============================================================

common_cols = [
    "source", "pmid", "doi", "eid", "title", "abstract", "year",
    "journal", "authors", "affiliations", "keywords",
    "language", "document_type"
]

for df_ in [pubmed_df, scopus_df]:
    for c in common_cols:
        if c not in df_.columns:
            df_[c] = ""

merged_df = pd.concat(
    [pubmed_df[common_cols], scopus_df[common_cols]],
    ignore_index=True
)

merged_df["title_clean"] = merged_df["title"].apply(normalize_title)
merged_df["doi_clean"] = merged_df["doi"].apply(normalize_doi)
merged_df["pmid_clean"] = merged_df["pmid"].fillna("").astype(str).str.strip()

save_csv(merged_df, "05_merged_raw_pubmed_scopus.csv")

# ============================================================
# CELL 10. DEDUPLICATION WITH LOGGING
# ============================================================

dedup_log = []

n0 = len(merged_df)

# DOI deduplication
before = len(merged_df)
doi_nonempty = merged_df["doi_clean"] != ""
df_doi = merged_df[doi_nonempty].drop_duplicates("doi_clean", keep="first")
df_no_doi = merged_df[~doi_nonempty]
merged_df = pd.concat([df_doi, df_no_doi], ignore_index=True)
dedup_log.append(["Removed duplicate DOI records", before - len(merged_df)])

# PMID deduplication
before = len(merged_df)
pmid_nonempty = merged_df["pmid_clean"] != ""
df_pmid = merged_df[pmid_nonempty].drop_duplicates("pmid_clean", keep="first")
df_no_pmid = merged_df[~pmid_nonempty]
merged_df = pd.concat([df_pmid, df_no_pmid], ignore_index=True)
dedup_log.append(["Removed duplicate PMID records", before - len(merged_df)])

# Exact title deduplication
before = len(merged_df)
merged_df = merged_df.drop_duplicates("title_clean", keep="first")
dedup_log.append(["Removed exact title duplicates", before - len(merged_df)])

# Fuzzy title deduplication
before = len(merged_df)
merged_df = merged_df.reset_index(drop=True)

to_remove = set()
titles = merged_df["title_clean"].tolist()

for i in tqdm(range(len(titles))):
    if i in to_remove:
        continue
    ti = titles[i]
    if not ti or len(ti) < 20:
        continue

    for j in range(i + 1, len(titles)):
        if j in to_remove:
            continue
        tj = titles[j]
        if not tj or len(tj) < 20:
            continue

        if fuzz.ratio(ti, tj) >= TITLE_SIM_THRESHOLD:
            to_remove.add(j)

merged_df = merged_df.drop(index=list(to_remove)).reset_index(drop=True)
dedup_log.append(["Removed fuzzy title duplicates", before - len(merged_df)])

dedup_df = merged_df.copy()
save_csv(dedup_df, "06_deduplicated_records.csv")

dedup_log_df = pd.DataFrame(dedup_log, columns=["step", "records_removed"])
save_csv(dedup_log_df, "07_deduplication_log.csv")

print("Initial merged:", n0)
print("After deduplication:", len(dedup_df))

# ============================================================
# CELL 11. FILTERING AND FINAL NLP CORPUS
# ============================================================

filter_log = []

df = dedup_df.copy()
before = len(df)

df = df[df["title"].fillna("").str.len() > 5]
filter_log.append(["Removed records with missing/short title", before - len(df)])

before = len(df)
df = df[df["year"].notna()]
filter_log.append(["Removed records with missing year", before - len(df)])

before = len(df)
df["year"] = df["year"].astype(int)
df = df[(df["year"] >= YEAR_START) & (df["year"] <= YEAR_END)]
filter_log.append(["Removed records outside year range", before - len(df)])

if REQUIRE_ABSTRACT_FOR_NLP:
    before = len(df)
    df = df[df["abstract"].fillna("").str.len() >= 30]
    filter_log.append(["Removed records without usable abstract", before - len(df)])
else:
    filter_log.append(["Abstract not mandatory for NLP corpus", 0])

df["text_for_nlp"] = (
    df["title"].fillna("") + ". " +
    df["abstract"].fillna("") + ". " +
    df["keywords"].fillna("")
).apply(clean_text)

before = len(df)
df = df[df["text_for_nlp"].str.len() >= 30]
filter_log.append(["Removed records with insufficient combined NLP text", before - len(df)])

final_corpus = df.reset_index(drop=True)
save_csv(final_corpus, "08_final_nlp_corpus.csv")

filter_log_df = pd.DataFrame(filter_log, columns=["step", "records_removed"])
save_csv(filter_log_df, "09_filtering_log.csv")

final_corpus.head()

# ============================================================
# CELL 12. PRISMA COUNTS
# ============================================================

prisma_rows = [
    ["Identified from PubMed", len(pubmed_df)],
    ["Identified from Scopus", len(scopus_df)],
    ["Identified total before deduplication", n0],
    ["Removed duplicate DOI records", int(dedup_log_df.loc[dedup_log_df["step"] == "Removed duplicate DOI records", "records_removed"].iloc[0])],
    ["Removed duplicate PMID records", int(dedup_log_df.loc[dedup_log_df["step"] == "Removed duplicate PMID records", "records_removed"].iloc[0])],
    ["Removed exact title duplicates", int(dedup_log_df.loc[dedup_log_df["step"] == "Removed exact title duplicates", "records_removed"].iloc[0])],
    ["Removed fuzzy title duplicates", int(dedup_log_df.loc[dedup_log_df["step"] == "Removed fuzzy title duplicates", "records_removed"].iloc[0])],
    ["Records after deduplication", len(dedup_df)],
    ["Final NLP corpus", len(final_corpus)]
]

for _, row in filter_log_df.iterrows():
    prisma_rows.append([row["step"], int(row["records_removed"])])

prisma_df = pd.DataFrame(prisma_rows, columns=["PRISMA_step", "count"])
save_csv(prisma_df, "10_prisma_counts.csv")
prisma_df

# ============================================================
# CELL 13. BERTopic MODELING
# ============================================================

texts = final_corpus["text_for_nlp"].astype(str).tolist()

embedding_model = SentenceTransformer("all-mpnet-base-v2")

umap_model = UMAP(
    n_neighbors=15,
    n_components=5,
    min_dist=0.0,
    metric="cosine",
    random_state=42
)

hdbscan_model = HDBSCAN(
    min_cluster_size=15,
    min_samples=5,
    metric="euclidean",
    prediction_data=True
)

topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    min_topic_size=15,
    calculate_probabilities=True,
    verbose=True
)

topics, probs = topic_model.fit_transform(texts)

final_corpus["topic_initial"] = topics
topic_info_initial = topic_model.get_topic_info()

save_csv(final_corpus, "11_corpus_with_initial_topics.csv")
save_csv(topic_info_initial, "12_topic_info_initial.csv")

topic_info_initial.head(20)

# ============================================================
# CELL 14. OUTLIER REDUCTION
# ============================================================

topics_refined = topic_model.reduce_outliers(texts, topics, strategy="embeddings")
final_corpus["topic_refined"] = topics_refined

topic_info_refined = topic_model.get_topic_info()
save_csv(final_corpus, "13_corpus_with_refined_topics.csv")
save_csv(topic_info_refined, "14_topic_info_refined.csv")

topic_info_refined.head(20)

# ============================================================
# CELL 15. TOPIC COHERENCE AND TOPIC DIVERSITY
# ============================================================

def simple_tokenize(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    tokens = [t for t in text.split() if len(t) > 3]
    stop = {
        "with", "from", "that", "this", "were", "have", "been", "their",
        "using", "study", "case", "cases", "mushroom", "poisoning",
        "poisoned", "poisonous", "patient", "patients"
    }
    return [t for t in tokens if t not in stop]

tokenized_texts = [simple_tokenize(t) for t in texts]
dictionary = Dictionary(tokenized_texts)

topic_words = []
valid_topics = [
    t for t in topic_model.get_topics().keys()
    if t != -1 and len(topic_model.get_topic(t)) > 0
]

for t in valid_topics:
    words = [w for w, _ in topic_model.get_topic(t)[:10]]
    topic_words.append(words)

coherence_cv = np.nan
if len(topic_words) > 0:
    coherence_model = CoherenceModel(
        topics=topic_words,
        texts=tokenized_texts,
        dictionary=dictionary,
        coherence="c_v"
    )
    coherence_cv = coherence_model.get_coherence()

all_top_words = [w for topic in topic_words for w in topic]
topic_diversity = len(set(all_top_words)) / len(all_top_words) if all_top_words else np.nan

coherence_df = pd.DataFrame({
    "metric": ["c_v_topic_coherence", "topic_diversity", "number_of_valid_topics"],
    "value": [coherence_cv, topic_diversity, len(valid_topics)]
})

save_csv(coherence_df, "15_topic_validation_metrics.csv")
coherence_df

# ============================================================
# CELL 16. REPRESENTATIVE DOCUMENTS FOR EXPERT REVIEW
# ============================================================

review_rows = []

for t in valid_topics:
    rep_docs = topic_model.get_representative_docs(t)
    words = ", ".join([w for w, _ in topic_model.get_topic(t)[:10]])

    if isinstance(rep_docs, dict):
        docs = rep_docs.get(t, [])
    else:
        docs = rep_docs

    for i, doc in enumerate(docs[:5]):
        review_rows.append({
            "topic_id": t,
            "top_words": words,
            "representative_doc_rank": i + 1,
            "representative_text": doc[:600],
            "expert_label": "",
            "include_exclude": "",
            "comment": ""
        })

expert_review_df = pd.DataFrame(review_rows)
save_csv(expert_review_df, "16_topic_expert_review_template.csv")
expert_review_df.head()

# ============================================================
# CELL 17. MANUAL TOPIC MAPPING
# Edit this after reading 16_topic_expert_review_template.csv
# ============================================================

topic_mapping = {
    -1: "Unassigned / mixed",

    # Update these after inspecting actual topic IDs.
    # These are example thematic bins expected in mushroom-poisoning literature.
    0: "Clinical poisoning cases and diagnosis",
    1: "Liver failure and hepatotoxicity",
    2: "Poisonous versus edible mushroom studies",
    3: "Outbreaks and epidemiology",
    4: "Toxin detection and biomonitoring",
    5: "Amanita and amatoxin poisoning",
    6: "AI and machine-learning classification",
    7: "Renal toxicity and Cortinarius poisoning",
    8: "Experimental hepatotoxicity models",
    9: "Neurological toxicity and encephalopathy",
    10: "Shiitake dermatitis"
}

final_corpus["final_topic"] = (
    final_corpus["topic_refined"]
    .map(topic_mapping)
    .fillna("Other mushroom-poisoning topics")
)

save_csv(final_corpus, "17_final_dataset_with_topics.csv")

topic_summary_final = (
    final_corpus["final_topic"]
    .value_counts()
    .reset_index()
)

topic_summary_final.columns = ["final_topic", "count"]
topic_summary_final["percent"] = (
    topic_summary_final["count"] / topic_summary_final["count"].sum() * 100
).round(2)

save_csv(topic_summary_final, "18_final_topic_summary.csv")
topic_summary_final

# ============================================================
# CELL 18. TOPIC DISTRIBUTION FIGURES
# ============================================================

plt.figure(figsize=(12, 7))
plt.bar(topic_summary_final["final_topic"], topic_summary_final["count"])
plt.ylabel("Number of publications")
plt.xlabel("Final topic")
plt.title("Distribution of mushroom-poisoning research themes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("outputs/figures/figure_topic_distribution.png", dpi=300)
plt.show()

top5 = topic_summary_final.head(5)

plt.figure(figsize=(8, 6))
plt.bar(top5["final_topic"], top5["count"])
plt.ylabel("Number of publications")
plt.title("Top five mushroom-poisoning research themes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("outputs/figures/figure_top5_topics.png", dpi=300)
plt.show()

# ============================================================
# CELL 19. TEMPORAL TREND ANALYSIS
# ============================================================

df_trend = final_corpus.copy()
df_trend = df_trend[df_trend["final_topic"] != "Unassigned / mixed"]
df_trend = df_trend[df_trend["year"].notna()]
df_trend["year"] = df_trend["year"].astype(int)

trend_long = (
    df_trend.groupby(["year", "final_topic"])
    .size()
    .reset_index(name="count")
)

trend_wide = (
    trend_long
    .pivot(index="year", columns="final_topic", values="count")
    .fillna(0)
)

save_csv(trend_long, "19_temporal_trends_long.csv")
save_csv(trend_wide.reset_index(), "20_temporal_trends_wide.csv")

plt.figure(figsize=(14, 7))
for col in trend_wide.columns:
    plt.plot(trend_wide.index, trend_wide[col], linewidth=1.5, label=col)

plt.xlabel("Year")
plt.ylabel("Number of publications")
plt.title("Temporal evolution of mushroom-poisoning research themes")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=8)
plt.tight_layout()
plt.savefig("outputs/figures/figure_temporal_trends_all_topics.png", dpi=300)
plt.show()

# ============================================================
# CELL 20. TOP THEME TEMPORAL TRENDS
# ============================================================

top_topics = topic_summary_final.head(6)["final_topic"].tolist()
top_topics = [t for t in top_topics if t in trend_wide.columns]

trend_top = trend_wide[top_topics]

plt.figure(figsize=(12, 6))
for col in trend_top.columns:
    plt.plot(trend_top.index, trend_top[col], linewidth=2, label=col)

plt.xlabel("Year")
plt.ylabel("Number of publications")
plt.title("Temporal trends of major mushroom-poisoning themes")
plt.legend()
plt.tight_layout()
plt.savefig("outputs/figures/figure_temporal_trends_top6.png", dpi=300)
plt.show()

trend_smooth = trend_top.rolling(window=3, center=True, min_periods=1).mean()

plt.figure(figsize=(12, 6))
for col in trend_smooth.columns:
    plt.plot(trend_smooth.index, trend_smooth[col], linewidth=2, label=col)

plt.xlabel("Year")
plt.ylabel("Publications, 3-year smoothed")
plt.title("Smoothed temporal trends of major themes")
plt.legend()
plt.tight_layout()
plt.savefig("outputs/figures/figure_temporal_trends_smoothed.png", dpi=300)
plt.show()

trend_pct = trend_wide.div(trend_wide.sum(axis=1), axis=0) * 100
trend_pct_top = trend_pct[top_topics]

plt.figure(figsize=(12, 6))
for col in trend_pct_top.columns:
    plt.plot(trend_pct_top.index, trend_pct_top[col], linewidth=2, label=col)

plt.xlabel("Year")
plt.ylabel("Percentage of annual publications")
plt.title("Relative contribution of major themes over time")
plt.legend()
plt.tight_layout()
plt.savefig("outputs/figures/figure_temporal_trends_percentage.png", dpi=300)
plt.show()

save_csv(trend_top.reset_index(), "21_top_theme_trends.csv")
save_csv(trend_smooth.reset_index(), "22_smoothed_top_theme_trends.csv")
save_csv(trend_pct.reset_index(), "23_relative_topic_contribution.csv")

# ============================================================
# CELL 21. REGRESSION TREND STATISTICS
# ============================================================

trend_stats = []

for topic in trend_top.columns:
    x = trend_top.index.values.astype(float)
    y = trend_top[topic].values.astype(float)

    if len(x) >= 3 and np.sum(y) > 0:
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
    else:
        slope, intercept, r_value, p_value, std_err = [np.nan] * 5

    trend_stats.append({
        "topic": topic,
        "slope_publications_per_year": slope,
        "r_squared": r_value ** 2 if not pd.isna(r_value) else np.nan,
        "p_value": p_value,
        "std_err": std_err
    })

trend_stats_df = pd.DataFrame(trend_stats).sort_values("p_value")
save_csv(trend_stats_df, "24_trend_regression_statistics.csv")
trend_stats_df

# ============================================================
# CELL 22. GROWTH RATE ANALYSIS
# ============================================================

growth_rows = []

for topic in trend_top.columns:
    start_count = trend_top[topic].iloc[0]
    end_count = trend_top[topic].iloc[-1]

    growth_percent = ((end_count - start_count) / max(start_count, 1)) * 100

    growth_rows.append({
        "topic": topic,
        "start_year": int(trend_top.index.min()),
        "end_year": int(trend_top.index.max()),
        "start_count": start_count,
        "end_count": end_count,
        "growth_percent": growth_percent
    })

growth_df = pd.DataFrame(growth_rows).sort_values("growth_percent", ascending=False)
save_csv(growth_df, "25_topic_growth_rates.csv")
growth_df

# ============================================================
# CELL 23. SHANNON DIVERSITY INDEX
# ============================================================

p = topic_summary_final["percent"] / 100
shannon_index = -np.sum(p * np.log(p))

shannon_df = pd.DataFrame({
    "metric": ["Shannon diversity index"],
    "value": [shannon_index]
})

save_csv(shannon_df, "26_topic_shannon_diversity.csv")
shannon_df

# ============================================================
# CELL 24. SPECIES / TOXIN KEYWORD ANALYSIS
# ============================================================

species_terms = {
    "Amanita phalloides": r"amanita phalloides",
    "Amanita virosa": r"amanita virosa",
    "Amanita muscaria": r"amanita muscaria",
    "Galerina": r"\bgalerina\b",
    "Cortinarius": r"\bcortinarius\b",
    "Gyromitra": r"\bgyromitra\b",
    "Lepiota": r"\blepiota\b",
    "Inocybe": r"\binocybe\b",
    "Psilocybe": r"\bpsilocybe\b",
    "Amatoxin": r"\bamatoxin|\bamatoxins",
    "Amanitin": r"\bamanitin|\bamanitins",
    "Orellanine": r"\borellanine\b",
    "Muscarine": r"\bmuscarine\b",
    "Ibotenic acid": r"ibotenic acid",
    "Muscimol": r"\bmuscimol\b"
}

species_rows = []

for label, pattern in species_terms.items():
    mask = final_corpus["text_for_nlp"].str.lower().str.contains(pattern, regex=True, na=False)
    species_rows.append({
        "term": label,
        "count": int(mask.sum()),
        "percent": round(mask.mean() * 100, 2)
    })

species_df = pd.DataFrame(species_rows).sort_values("count", ascending=False)
save_csv(species_df, "27_species_toxin_keyword_frequency.csv")

plt.figure(figsize=(10, 6))
plt.bar(species_df["term"], species_df["count"])
plt.ylabel("Number of records")
plt.title("Frequency of major mushroom species and toxin terms")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("outputs/figures/figure_species_toxin_frequency.png", dpi=300)
plt.show()

species_df

# ============================================================
# CELL 25. COUNTRY EXTRACTION FROM AFFILIATIONS
# ============================================================

country_patterns = {
    "United States": ["united states", "usa", "u.s.a", "u.s."],
    "China": ["china"],
    "Japan": ["japan"],
    "India": ["india"],
    "Turkey": ["turkey", "türkiye"],
    "Iran": ["iran"],
    "France": ["france"],
    "Germany": ["germany"],
    "Italy": ["italy"],
    "Poland": ["poland"],
    "Spain": ["spain"],
    "United Kingdom": ["united kingdom", "uk", "england", "scotland", "wales"],
    "South Korea": ["south korea", "korea"],
    "Canada": ["canada"],
    "Australia": ["australia"],
    "Brazil": ["brazil"],
    "Mexico": ["mexico"],
    "Russia": ["russia"],
    "Czech Republic": ["czech"],
    "Portugal": ["portugal"],
    "Switzerland": ["switzerland"],
    "Sweden": ["sweden"],
    "Norway": ["norway"],
    "Netherlands": ["netherlands"],
    "Greece": ["greece"]
}

def extract_countries(text):
    text = clean_text(text).lower()
    found = []
    for country, patterns in country_patterns.items():
        for ptn in patterns:
            if re.search(r"\b" + re.escape(ptn) + r"\b", text):
                found.append(country)
                break
    return sorted(set(found))

country_rows = []

for _, row in final_corpus.iterrows():
    aff_text = " ".join([
        clean_text(row.get("affiliations", "")),
        clean_text(row.get("authors", "")),
        clean_text(row.get("journal", ""))
    ])

    countries = extract_countries(aff_text)

    for c in countries:
        country_rows.append({
            "title": row["title"],
            "year": row["year"],
            "final_topic": row["final_topic"],
            "country": c
        })

country_df = pd.DataFrame(country_rows)

if len(country_df) > 0:
    country_count = (
        country_df["country"]
        .value_counts()
        .reset_index()
    )
    country_count.columns = ["country", "count"]
else:
    country_count = pd.DataFrame(columns=["country", "count"])

save_csv(country_df, "28_country_record_mapping.csv")
save_csv(country_count, "29_country_publication_counts.csv")

country_count.head(20)

# ============================================================
# CELL 26. WOS VALIDATION COMPARISON
# ============================================================

if len(wos_df) > 0:
    wos_df["text_for_nlp"] = (
        wos_df["title"].fillna("") + ". " +
        wos_df["abstract"].fillna("") + ". " +
        wos_df["keywords"].fillna("")
    ).apply(clean_text)

    wos_df["year"] = pd.to_numeric(wos_df["year"], errors="coerce")

    wos_summary = pd.DataFrame({
        "metric": [
            "WoS validation records retrieved",
            "WoS records with DOI",
            "WoS records with abstract",
            "WoS year min",
            "WoS year max"
        ],
        "value": [
            len(wos_df),
            int((wos_df["doi"].fillna("") != "").sum()),
            int((wos_df["abstract"].fillna("").str.len() > 30).sum()),
            wos_df["year"].min(),
            wos_df["year"].max()
        ]
    })
else:
    wos_summary = pd.DataFrame({
        "metric": ["WoS validation records retrieved"],
        "value": [0]
    })

save_csv(wos_summary, "30_wos_validation_summary.csv")
wos_summary

# ============================================================
# CELL 27. PRISMA FIGURE
# ============================================================

prisma_dict = dict(zip(prisma_df["PRISMA_step"], prisma_df["count"]))

identified_pubmed = prisma_dict.get("Identified from PubMed", 0)
identified_scopus = prisma_dict.get("Identified from Scopus", 0)
identified_total = prisma_dict.get("Identified total before deduplication", 0)
after_dedup = prisma_dict.get("Records after deduplication", 0)
final_nlp = prisma_dict.get("Final NLP corpus", 0)

duplicates_removed = identified_total - after_dedup
records_removed_filtering = after_dedup - final_nlp

plt.figure(figsize=(10, 10))
ax = plt.gca()
ax.axis("off")

def box(x, y, text):
    ax.text(
        x, y, text,
        ha="center", va="center",
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="black"),
        fontsize=11
    )

def arrow(y1, y2):
    ax.annotate(
        "", xy=(0.5, y2), xytext=(0.5, y1),
        arrowprops=dict(arrowstyle="->", lw=1.5)
    )

box(0.5, 0.90, f"Records identified\nPubMed: {identified_pubmed}\nScopus: {identified_scopus}\nTotal: {identified_total}")
arrow(0.84, 0.75)
box(0.5, 0.70, f"Records after deduplication\nn = {after_dedup}\nDuplicates removed = {duplicates_removed}")
arrow(0.64, 0.55)
box(0.5, 0.50, f"Records screened for NLP eligibility\nn = {after_dedup}")
arrow(0.44, 0.35)
box(0.5, 0.30, f"Final corpus included in NLP analysis\nn = {final_nlp}\nFiltering exclusions = {records_removed_filtering}")

plt.title("PRISMA-style literature selection workflow", fontsize=14)
plt.tight_layout()
plt.savefig("outputs/figures/figure_prisma_workflow.png", dpi=300)
plt.show()

# ============================================================
# CELL 28. SAVE BERTopic INTERACTIVE FIGURES
# ============================================================

try:
    fig_topics = topic_model.visualize_topics()
    fig_topics.write_html("outputs/figures/interactive_topic_map.html")

    fig_barchart = topic_model.visualize_barchart(top_n_topics=15)
    fig_barchart.write_html("outputs/figures/interactive_topic_barchart.html")

    topics_over_time = topic_model.topics_over_time(
        texts,
        final_corpus["year"].fillna(0).astype(int).tolist()
    )

    fig_time = topic_model.visualize_topics_over_time(topics_over_time)
    fig_time.write_html("outputs/figures/interactive_topics_over_time.html")

    print("Interactive BERTopic figures saved.")

except Exception as e:
    print("Interactive figure generation skipped:", e)

# ============================================================
# CELL 29. MANUSCRIPT METHODS SUMMARY TABLE
# ============================================================

methods_summary = pd.DataFrame({
    "component": [
        "Databases",
        "Search period",
        "Primary search concepts",
        "Deduplication",
        "NLP model",
        "Embedding model",
        "Dimensionality reduction",
        "Clustering",
        "Topic validation",
        "Trend analysis",
        "Diversity metric",
        "Validation subset"
    ],
    "description": [
        "PubMed and Scopus; Web of Science Starter API used as supplementary validation subset",
        f"{YEAR_START}–{YEAR_END}",
        "mushroom poisoning, mushroom intoxication, poisonous mushroom, mycetism, amatoxin, amanitin, mushroom toxin, Amanita poisoning",
        "Sequential DOI, PMID, exact-title, and fuzzy-title deduplication",
        "BERTopic",
        "all-mpnet-base-v2 SentenceTransformer",
        "UMAP",
        "HDBSCAN",
        "c_v topic coherence, topic diversity, representative-document review template",
        "Annual publication counts, percentage contribution, 3-year smoothing, linear regression",
        "Shannon diversity index",
        "WoS records summarized separately to assess external consistency"
    ]
})

save_csv(methods_summary, "31_methods_summary_for_manuscript.csv")
methods_summary

# ============================================================
# CELL 30. LIMITATIONS TEXT FOR MANUSCRIPT
# ============================================================

limitations_text = """
Key methodological limitations:
1. Scopus and PubMed differ in indexing structure, metadata completeness, and abstract availability.
2. Web of Science Starter API was used as a supplementary validation subset rather than as a full analytical corpus.
3. Topic labels were generated computationally and require domain-expert review before final interpretation.
4. Country-level mapping based on affiliation strings may undercount multinational collaborations and records with incomplete metadata.
5. BERTopic results may vary with embedding model, UMAP, and HDBSCAN parameters; therefore, topic coherence and representative-document review were included.
6. Records without sufficient title/abstract/keyword text may be underrepresented in NLP-based analyses.
"""

with open("outputs/tables/32_limitations_text_for_manuscript.txt", "w") as f:
    f.write(limitations_text)

print(limitations_text)

# ============================================================
# CELL 31. ZIP ALL OUTPUTS
# ============================================================

zip_path = "mushroom_poisoning_nlp_publication_outputs.zip"

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    for root, dirs, file_list in os.walk("outputs"):
        for file in file_list:
            full_path = os.path.join(root, file)
            z.write(full_path)

print(f"Created: {zip_path}")
files.download(zip_path)
```
