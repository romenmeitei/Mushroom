```python
# ============================================================
# FINAL THEME CONSOLIDATION + MANUSCRIPT FIGURES
# For mushroom poisoning BERTopic manuscript
# ============================================================

!pip -q install pandas numpy matplotlib seaborn openpyxl plotly kaleido rapidfuzz

import os
import re
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

os.makedirs("final_theme_outputs", exist_ok=True)
os.makedirs("final_theme_outputs/tables", exist_ok=True)
os.makedirs("final_theme_outputs/figures", exist_ok=True)

# ============================================================
# 1. UPLOAD REQUIRED FILES
# ============================================================

uploaded = files.upload()

# Required files:
# 1. 16_topic_expert_review_template_annotated.csv
# 2. 17_final_dataset_with_topics.csv

annotated_file = "16_topic_expert_review_template_annotated.csv"
dataset_file = "17_final_dataset_with_topics.csv"

expert_df = pd.read_csv(annotated_file)
corpus_df = pd.read_csv(dataset_file)

print("Expert annotation file:", expert_df.shape)
print("Corpus file:", corpus_df.shape)

display(expert_df.head())
display(corpus_df.head())

# ============================================================
# 2. CLEAN COLUMN NAMES
# ============================================================

expert_df.columns = expert_df.columns.str.strip()
corpus_df.columns = corpus_df.columns.str.strip()

print("Expert columns:", expert_df.columns.tolist())
print("Corpus columns:", corpus_df.columns.tolist())

# ============================================================
# 3. CREATE TOPIC-LEVEL CURATION SUMMARY
# ============================================================

topic_summary = (
    expert_df
    .groupby("topic_id")
    .agg({
        "top_words": "first",
        "expert_label": "first",
        "include_exclude": "first",
        "comment": "first"
    })
    .reset_index()
)

topic_summary["topic_id"] = topic_summary["topic_id"].astype(int)

topic_summary.to_csv(
    "final_theme_outputs/tables/01_topic_level_expert_curation_summary.csv",
    index=False
)

display(topic_summary.head(20))

# ============================================================
# 4. DEFINE FINAL CONSOLIDATED THEMES
# ============================================================
# This is the critical manuscript-level mapping.
# Edit this manually after reviewing your annotated CSV.
# The topic IDs below are examples based on typical mushroom poisoning themes.
# Adjust topic IDs if your expert_label differs.

def assign_final_theme(row):
    label = str(row["expert_label"]).lower()
    decision = str(row["include_exclude"]).upper()

    if decision == "EXCLUDE":
        return "Excluded / non-core topic"

    if "amanita" in label or "amatoxin" in label or "amanitin" in label:
        if "liver" in label or "hepato" in label or "failure" in label:
            return "Amatoxin-induced hepatotoxicity and liver failure"
        else:
            return "Amanita poisoning and amatoxin intoxication"

    if "liver" in label or "hepatic" in label or "hepatotoxic" in label:
        return "Amatoxin-induced hepatotoxicity and liver failure"

    if "clinical" in label or "management" in label or "emergency" in label or "supportive" in label:
        return "Clinical poisoning management and supportive care"

    if "diagnosis" in label or "diagnostic" in label:
        return "Diagnosis, biomarkers, and clinical recognition"

    if "toxin detection" in label or "analytical" in label or "lc-ms" in label or "mass spectrometry" in label:
        return "Toxin detection and analytical toxicology"

    if "taxonomy" in label or "identification" in label or "species" in label or "barcoding" in label:
        return "Poisonous mushroom identification and taxonomy"

    if "epidemiology" in label or "outbreak" in label or "surveillance" in label:
        return "Epidemiology, outbreaks, and surveillance"

    if "renal" in label or "kidney" in label or "nephro" in label or "orellanine" in label or "cortinarius" in label:
        return "Renal and extrahepatic toxicity"

    if "neurolog" in label or "muscimol" in label or "ibotenic" in label or "muscarine" in label:
        return "Neurotoxic and cholinergic mushroom poisoning"

    if "ai" in label or "machine" in label or "deep learning" in label or "computational" in label or "image" in label:
        return "AI-assisted mushroom identification and computational classification"

    if "public health" in label or "prevention" in label or "awareness" in label:
        return "Public health prevention and risk communication"

    if "experimental" in label or "animal" in label or "model" in label:
        return "Experimental toxicology and translational models"

    if "therapy" in label or "treatment" in label or "n-acetylcysteine" in label or "silibinin" in label:
        return "Therapeutic strategies and antidotal interventions"

    if decision == "PARTIAL":
        return "Partially relevant mushroom-toxicology topic"

    return "Other mushroom-poisoning research themes"


topic_summary["final_theme"] = topic_summary.apply(assign_final_theme, axis=1)

topic_summary.to_csv(
    "final_theme_outputs/tables/02_final_theme_mapping_table.csv",
    index=False
)

display(topic_summary)

# ============================================================
# 5. MERGE FINAL THEMES INTO FULL CORPUS
# ============================================================

# corpus_df should contain topic_refined from previous pipeline.
# If your column name differs, adjust here.

if "topic_refined" in corpus_df.columns:
    topic_col = "topic_refined"
elif "topic_initial" in corpus_df.columns:
    topic_col = "topic_initial"
else:
    raise ValueError("No topic_refined or topic_initial column found in corpus file.")

corpus_df[topic_col] = corpus_df[topic_col].astype(int)

theme_map = dict(zip(topic_summary["topic_id"], topic_summary["final_theme"]))
decision_map = dict(zip(topic_summary["topic_id"], topic_summary["include_exclude"]))
label_map = dict(zip(topic_summary["topic_id"], topic_summary["expert_label"]))

corpus_df["expert_label"] = corpus_df[topic_col].map(label_map)
corpus_df["include_exclude"] = corpus_df[topic_col].map(decision_map)
corpus_df["final_theme"] = corpus_df[topic_col].map(theme_map)

corpus_df["final_theme"] = corpus_df["final_theme"].fillna("Unmapped topic")
corpus_df["include_exclude"] = corpus_df["include_exclude"].fillna("UNMAPPED")

corpus_df.to_csv(
    "final_theme_outputs/tables/03_full_corpus_with_final_themes.csv",
    index=False
)

display(corpus_df[["title", "year", topic_col, "expert_label", "include_exclude", "final_theme"]].head())

# ============================================================
# 6. FILTER INCLUDED THEMES ONLY
# ============================================================

included_df = corpus_df[
    corpus_df["include_exclude"].astype(str).str.upper().isin(["INCLUDE", "PARTIAL"])
].copy()

included_df = included_df[
    included_df["final_theme"] != "Excluded / non-core topic"
].copy()

included_df.to_csv(
    "final_theme_outputs/tables/04_included_corpus_final_themes_only.csv",
    index=False
)

print("Full corpus:", corpus_df.shape)
print("Included mushroom-specific corpus:", included_df.shape)

# ============================================================
# 7. FINAL THEME COUNTS
# ============================================================

theme_counts = (
    included_df["final_theme"]
    .value_counts()
    .reset_index()
)

theme_counts.columns = ["final_theme", "count"]
theme_counts["percent"] = (
    theme_counts["count"] / theme_counts["count"].sum() * 100
).round(2)

theme_counts.to_csv(
    "final_theme_outputs/tables/05_final_theme_counts.csv",
    index=False
)

display(theme_counts)

# ============================================================
# 8. FINAL THEME DISTRIBUTION FIGURE
# ============================================================

plt.figure(figsize=(12, 7))
plt.bar(theme_counts["final_theme"], theme_counts["count"])
plt.ylabel("Number of publications")
plt.xlabel("Final expert-curated theme")
plt.title("Distribution of expert-curated mushroom-poisoning research themes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_3_final_theme_distribution.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_3_final_theme_distribution.pdf")
plt.show()

# ============================================================
# 9. YEARLY FINAL THEME TRENDS
# ============================================================

included_df["year"] = pd.to_numeric(included_df["year"], errors="coerce")
included_df = included_df.dropna(subset=["year"])
included_df["year"] = included_df["year"].astype(int)

trend_long = (
    included_df
    .groupby(["year", "final_theme"])
    .size()
    .reset_index(name="count")
)

trend_wide = (
    trend_long
    .pivot(index="year", columns="final_theme", values="count")
    .fillna(0)
)

trend_long.to_csv(
    "final_theme_outputs/tables/06_final_theme_temporal_trends_long.csv",
    index=False
)

trend_wide.reset_index().to_csv(
    "final_theme_outputs/tables/07_final_theme_temporal_trends_wide.csv",
    index=False
)

display(trend_wide.head())

# ============================================================
# 10. TEMPORAL TREND FIGURE — ALL FINAL THEMES
# ============================================================

plt.figure(figsize=(14, 8))
for col in trend_wide.columns:
    plt.plot(trend_wide.index, trend_wide[col], linewidth=1.8, label=col)

plt.xlabel("Year")
plt.ylabel("Number of publications")
plt.title("Temporal evolution of expert-curated mushroom-poisoning themes")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=8)
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_4_temporal_trends_all_final_themes.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_4_temporal_trends_all_final_themes.pdf")
plt.show()

# ============================================================
# 11. TOP 6 FINAL THEMES TREND FIGURE
# ============================================================

top_themes = theme_counts.head(6)["final_theme"].tolist()
trend_top = trend_wide[top_themes]

plt.figure(figsize=(12, 7))
for col in trend_top.columns:
    plt.plot(trend_top.index, trend_top[col], linewidth=2.2, marker="o", label=col)

plt.xlabel("Year")
plt.ylabel("Number of publications")
plt.title("Temporal trends of leading mushroom-poisoning themes")
plt.legend(fontsize=8)
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_5_top_final_theme_trends.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_5_top_final_theme_trends.pdf")
plt.show()

# ============================================================
# 12. 3-YEAR SMOOTHED TEMPORAL TRENDS
# ============================================================

trend_smooth = trend_top.rolling(window=3, center=True, min_periods=1).mean()

plt.figure(figsize=(12, 7))
for col in trend_smooth.columns:
    plt.plot(trend_smooth.index, trend_smooth[col], linewidth=2.5, label=col)

plt.xlabel("Year")
plt.ylabel("Publications, 3-year smoothed")
plt.title("Smoothed temporal evolution of leading mushroom-poisoning themes")
plt.legend(fontsize=8)
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_6_smoothed_final_theme_trends.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_6_smoothed_final_theme_trends.pdf")
plt.show()

trend_smooth.reset_index().to_csv(
    "final_theme_outputs/tables/08_smoothed_final_theme_trends.csv",
    index=False
)

# ============================================================
# 13. RELATIVE CONTRIBUTION OF FINAL THEMES
# ============================================================

trend_pct = trend_wide.div(trend_wide.sum(axis=1), axis=0) * 100
trend_pct = trend_pct.fillna(0)

trend_pct.reset_index().to_csv(
    "final_theme_outputs/tables/09_final_theme_relative_contribution.csv",
    index=False
)

plt.figure(figsize=(12, 7))
for col in trend_pct[top_themes].columns:
    plt.plot(trend_pct.index, trend_pct[col], linewidth=2.2, label=col)

plt.xlabel("Year")
plt.ylabel("Percentage of annual publications")
plt.title("Relative annual contribution of leading mushroom-poisoning themes")
plt.legend(fontsize=8)
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_7_relative_theme_contribution.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_7_relative_theme_contribution.pdf")
plt.show()

# ============================================================
# 14. REGRESSION TREND STATISTICS
# ============================================================

from scipy.stats import linregress

trend_stats = []

for theme in trend_wide.columns:
    x = trend_wide.index.values.astype(float)
    y = trend_wide[theme].values.astype(float)

    if len(x) >= 3 and np.sum(y) > 0:
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        r_squared = r_value ** 2
    else:
        slope, intercept, r_squared, p_value, std_err = [np.nan] * 5

    trend_stats.append({
        "final_theme": theme,
        "slope_publications_per_year": slope,
        "r_squared": r_squared,
        "p_value": p_value,
        "std_err": std_err
    })

trend_stats_df = pd.DataFrame(trend_stats)
trend_stats_df = trend_stats_df.sort_values("p_value")

trend_stats_df.to_csv(
    "final_theme_outputs/tables/10_final_theme_regression_statistics.csv",
    index=False
)

display(trend_stats_df)

# ============================================================
# 15. GROWTH RATE ANALYSIS
# ============================================================

growth_rows = []

for theme in trend_wide.columns:
    start_year = int(trend_wide.index.min())
    end_year = int(trend_wide.index.max())

    start_count = trend_wide.loc[start_year, theme]
    end_count = trend_wide.loc[end_year, theme]

    growth_percent = ((end_count - start_count) / max(start_count, 1)) * 100

    growth_rows.append({
        "final_theme": theme,
        "start_year": start_year,
        "end_year": end_year,
        "start_count": start_count,
        "end_count": end_count,
        "growth_percent": growth_percent
    })

growth_df = pd.DataFrame(growth_rows)
growth_df = growth_df.sort_values("growth_percent", ascending=False)

growth_df.to_csv(
    "final_theme_outputs/tables/11_final_theme_growth_rates.csv",
    index=False
)

display(growth_df)

# ============================================================
# 16. SHANNON DIVERSITY INDEX OF FINAL THEMES
# ============================================================

p = theme_counts["percent"] / 100
shannon_index = -np.sum(p * np.log(p))

diversity_df = pd.DataFrame({
    "metric": ["Shannon diversity index of final expert-curated themes"],
    "value": [shannon_index]
})

diversity_df.to_csv(
    "final_theme_outputs/tables/12_final_theme_shannon_diversity.csv",
    index=False
)

display(diversity_df)

# ============================================================
# 17. EXCLUDED TOPIC SUMMARY
# ============================================================

excluded_topics = topic_summary[
    topic_summary["include_exclude"].astype(str).str.upper() == "EXCLUDE"
].copy()

excluded_topics.to_csv(
    "final_theme_outputs/tables/13_excluded_topic_summary.csv",
    index=False
)

display(excluded_topics)

# ============================================================
# 18. INCLUDED / EXCLUDED PIE CHART
# ============================================================

decision_counts = (
    topic_summary["include_exclude"]
    .fillna("UNMAPPED")
    .value_counts()
    .reset_index()
)

decision_counts.columns = ["decision", "count"]

decision_counts.to_csv(
    "final_theme_outputs/tables/14_topic_inclusion_exclusion_counts.csv",
    index=False
)

plt.figure(figsize=(7, 7))
plt.pie(
    decision_counts["count"],
    labels=decision_counts["decision"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Expert curation decisions for BERTopic clusters")
plt.tight_layout()
plt.savefig("final_theme_outputs/figures/Figure_8_topic_curation_decisions.png", dpi=300)
plt.savefig("final_theme_outputs/figures/Figure_8_topic_curation_decisions.pdf")
plt.show()

# ============================================================
# 19. FINAL THEME MAPPING TABLE FOR MANUSCRIPT
# ============================================================

theme_mapping_table = (
    topic_summary
    .groupby("final_theme")
    .agg({
        "topic_id": lambda x: ", ".join(map(str, sorted(x))),
        "expert_label": lambda x: "; ".join(sorted(set(map(str, x)))),
        "include_exclude": "first",
        "top_words": lambda x: " | ".join(list(map(str, x))[:3])
    })
    .reset_index()
)

theme_mapping_table.columns = [
    "final_theme",
    "bertopic_topic_ids",
    "expert_interpretations",
    "curation_status",
    "representative_top_words"
]

theme_mapping_table.to_csv(
    "final_theme_outputs/tables/15_manuscript_final_theme_mapping_table.csv",
    index=False
)

display(theme_mapping_table)

# ============================================================
# 20. PUBLICATION-READY SUMMARY TABLE
# ============================================================

summary_table = theme_counts.merge(
    trend_stats_df,
    on="final_theme",
    how="left"
).merge(
    growth_df[["final_theme", "growth_percent"]],
    on="final_theme",
    how="left"
)

summary_table = summary_table[
    [
        "final_theme",
        "count",
        "percent",
        "slope_publications_per_year",
        "r_squared",
        "p_value",
        "growth_percent"
    ]
]

summary_table.to_csv(
    "final_theme_outputs/tables/16_publication_ready_final_theme_summary.csv",
    index=False
)

display(summary_table)

# ============================================================
# 21. MANUSCRIPT RESULTS TEXT AUTO-DRAFT
# ============================================================

top_theme = theme_counts.iloc[0]["final_theme"]
top_count = theme_counts.iloc[0]["count"]
top_percent = theme_counts.iloc[0]["percent"]

n_total = len(corpus_df)
n_included = len(included_df)
n_topics = topic_summary["topic_id"].nunique()
n_excluded = len(excluded_topics)
n_final_themes = theme_counts.shape[0]

results_text = f"""
Expert-guided thematic consolidation of BERTopic-derived clusters produced {n_final_themes} final mushroom-poisoning research themes.
From the full NLP corpus of {n_total} records, {n_included} records were retained within mushroom-specific or partially relevant expert-curated themes.
A total of {n_topics} machine-generated semantic clusters were reviewed, of which {n_excluded} were excluded from final thematic interpretation due to limited mushroom-specific relevance or non-core biomedical content.

The most represented final theme was "{top_theme}", comprising {top_count} records ({top_percent}% of the included expert-curated corpus).
Temporal trend analysis demonstrated differential growth patterns across themes, with annual publication trajectories quantified using linear regression and relative annual contribution analysis.
"""

with open("final_theme_outputs/tables/17_auto_drafted_results_text.txt", "w") as f:
    f.write(results_text)

print(results_text)

# ============================================================
# 22. ZIP FINAL OUTPUTS
# ============================================================

zip_path = "final_theme_manuscript_outputs.zip"

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    for root, dirs, file_list in os.walk("final_theme_outputs"):
        for file in file_list:
            full_path = os.path.join(root, file)
            z.write(full_path)

print("Created:", zip_path)
files.download(zip_path)
```
